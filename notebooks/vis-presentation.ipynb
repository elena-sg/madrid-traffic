```python
inputs = [20, 10, 5, 15, 30]
predictions = [32, 35, 36, 31, 12]

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(range(5), inputs, c='#2ca02c')
ax.plot(range(4, 10), [inputs[-1]] + predictions, c='#ff7f0e')
ax.scatter(range(5), inputs, edgecolors='k', label='Datos de entrada', c='#2ca02c', s=64)
ax.scatter(range(5, 10), predictions, marker='X', edgecolors='k', label='Predicción',
                  c='#ff7f0e', s=64)
ax.set_ylabel("Ocupación de la carretera")
ax.set_xlabel("Períodos de tiempo")
ax.legend(loc="upper left")
plt.savefig(r"C:\Users\Bened\PycharmProjects\madrid-traffic\figures\presentacion\short-term.png")
plt.show()
```

```python
from graph_traffic.config import data_path
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter

date_form = DateFormatter("%d-%m-%Y")
```

```python
df = pd.read_csv(f"{data_path}/03-by-location/traffic/1001.csv", parse_dates=["fecha"])
n = 100
l = 300
fig, ax = plt.subplots(figsize=(10, 4))
df = df.set_index("fecha")
ax.plot(df.iloc[n:n+l]["ocupacion"])
ax.xaxis.set_major_formatter(date_form)
ax.set_ylabel("Ocupación [%]")
ax.set_xlabel("Fecha")
plt.show()
```
